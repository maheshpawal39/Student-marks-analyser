```python
# Strip hidden spaces from column names
df.columns = df.columns.str.strip()

# Your original code will now work perfectly
print("Maths Topper")
print(df.loc[df["Maths"].idxmax(), ["Name", "Maths"]])

print("\nEnglish Topper")
print(df.loc[df["English"].idxmax(), ["Name", "English"]])

print("\nScience Topper")
print(df.loc[df["Science"].idxmax(), ["Name", "Science"]])
```

```python
def grade(avg):
    if avg >=85:
        return "A"
    elif avg >=70:
        return "B"
    else:
        return "C"
    
df["Grade"] = df["Average"].apply(grade)   
```
